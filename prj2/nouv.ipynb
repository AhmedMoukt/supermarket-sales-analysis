```python
import pandas as pd

df = pd.read_csv('supermarket_sales - Sheet1.csv')
df.head()
```

```python
df['City'].isnull().sum()
```

```python
df.duplicated()
```

```python
df.dropna()
```

```python
df.drop_duplicates()
```

```python
df.columns = [col.replace(' ', '_') if ' ' in col else col for col in df.columns]
```

```python
df
```

```python
# Chiffre d'affaires total
total_sales = df['Total'].sum()

# Nombre de transactions
nb_transactions = df.shape[0]

# Panier moyen
panier_moyen = df['Total'].mean()

# Chiffre d'affaires par ville
ca_par_ville = df.groupby('City')['Total'].sum()

# Chiffre d'affaires par type de produit
ca_par_produit = df.groupby('Product_line')['Total'].sum()

# Affichage des KPI
print("Chiffre d'affaires total :", total_sales)
print("Nombre de transactions :", nb_transactions)
print("Panier moyen :", panier_moyen)
print("Chiffre d'affaires par ville :\n", ca_par_ville)
print("Chiffre d'affaires par type de produit :\n", ca_par_produit)
```

```python
# ...existing code...
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Chiffre d'affaires par ville
plt.figure(figsize=(8,5))
sns.barplot(x=ca_par_ville.index, y=ca_par_ville.values, palette="viridis")
plt.title("Chiffre d'affaires par ville")
plt.ylabel("Chiffre d'affaires")
plt.xlabel("Ville")
plt.show()

# 2. Chiffre d'affaires par type de produit
plt.figure(figsize=(10,5))
sns.barplot(x=ca_par_produit.index, y=ca_par_produit.values, palette="magma")
plt.title("Chiffre d'affaires par type de produit")
plt.ylabel("Chiffre d'affaires")
plt.xlabel("Type de produit")
plt.xticks(rotation=45)
plt.show()
```
