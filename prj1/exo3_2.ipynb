```python
df_grouped = df.groupby(['job_title_short', 'job_country'])['salary_year_avg'].median().reset_index()

plt.figure(figsize=(14, 6))
sns.barplot(data=df_grouped, x='job_title_short', y='salary_year_avg', hue='job_country')
plt.xticks(rotation=45, ha='right')
plt.title('Salaire médian annuel par poste et pays')
plt.tight_layout()
plt.show()
```

```python
# Liste des pays que tu souhaites afficher
selected_countries = ['France', 'Germany', 'USA', 'Canada', 'UK']

# Filtrer les données pour ne garder que les pays sélectionnés
df_filtered = df[df['job_country'].isin(selected_countries)]

# Créer la pivot table avec les données filtrées
pivot_table = df_filtered.groupby(['job_title_short', 'job_country'])['salary_year_avg'].median().unstack()

# Visualiser la heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlGnBu", cbar_kws={'label': 'Salaire médian'})
plt.title("Salaire médian annuel par poste et pays (Sélectionnés)")
plt.xlabel("Pays")
plt.ylabel("Poste")
plt.show()
```

```python
# Définir les pays à inclure
selected_countries = ['France', 'Germany', 'USA', 'Canada', 'UK']

# Filtrer le DataFrame
filtered_df = df[df['job_country'].isin(selected_countries)]

# Calculer la médiane des salaires pour les pays sélectionnés
median_salaries = filtered_df.groupby('job_country')['salary_year_avg'].median().sort_values()

# Visualisation
plt.figure(figsize=(8, 5))
median_salaries.plot(kind='barh', color='skyblue')
plt.xlabel('Salaire médian annuel')
plt.title('Salaire médian annuel pour une sélection de pays')
plt.tight_layout()
plt.show()
```

```python

median_salaries.plot(kind='bar', color='orange')
plt.ylabel('Salaire médian annuel')
plt.xticks(rotation=45)
plt.title('Salaire médian annuel par pays')
plt.tight_layout()
plt.show()
```

```python
selected_countries = df[df['job_country'].isin(['France', 'Germany', 'UK', 'Italy'])]

# Calculer la médiane des salaires pour ces pays
median_salaries = selected_countries.groupby('job_country')['salary_year_avg'].median()

# Visualisation
plt.figure(figsize=(10, 6))
median_salaries.plot(kind='barh', color='lightgreen')
plt.xlabel('Salaire médian annuel')
plt.title('Salaire médian annuel pour certains pays')
plt.xticks(fontsize=10)
plt.yticks(fontsize=8)
plt.tight_layout()
plt.show()
```
