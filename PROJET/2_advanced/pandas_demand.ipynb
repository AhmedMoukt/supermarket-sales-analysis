```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df_us=df[df['job_country']=='United States'].copy()
```

```python
df_us
```

```python
df_us['job_posted_mounth']=df_us['job_posted_date'].dt.strftime('%B')
df_us
```

```python
df_us_pivot=df_us.pivot_table(index='job_posted_mounth',columns='job_title_short',aggfunc='size')
```

```python

df_us_pivot['mounth_no']=pd.to_datetime(df_us_pivot['job_posted_mounth'],format='%B').dt.month
df_us_pivot.sort_values('mounth_no',inplace=True)
df_us_pivot.sort_index('job_posted_mounth')
```

```python
df_us_pivot.reset_index(inplace=True)
```
